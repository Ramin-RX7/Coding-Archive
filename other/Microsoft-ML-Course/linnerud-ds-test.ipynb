```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
physiology = pd.read_csv(r"./datasets/linnerud_physiological.csv",sep=" ")
# print(physiology.head())
exercise = pd.read_csv(r"./datasets/linnerud_exercise.csv", sep=" ")
# print(exercise.head())

dataset = physiology.join(exercise)
print(dataset.head())
```

```python
plt.scatter(dataset["Waist"],dataset["Situps"])
plt.xlabel("Waist")
plt.ylabel("Situps")
plt.axvline(dataset["Waist"].mean(), color='magenta', linestyle='dashed', linewidth=2, )

print(dataset["Waist"].mean())
plt.show()
```

```python
# print(dataset["Situps"].array.reshape(1,-1))
# print(dataset["Situps"].array.reshape(-1,1))
```

```python
from sklearn import linear_model,model_selection
# dataset["Situps"].array

X_train, X_test, y_train, y_test = model_selection.train_test_split(
        dataset["Waist"] .array.reshape(-1,1),
        dataset["Situps"],
        test_size=0.33)
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test,  color='black')
# plt.scatter(X_train, y_train,  color='orange')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
# plt.scatter(dataset["Waist"],dataset["Situps"])

plt.xlabel('Waist Line')
plt.ylabel('Situps')
plt.title(' ')

plt.show()
```
